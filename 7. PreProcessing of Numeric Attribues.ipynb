# Normalization

Normalization means adjusting values measured on different scales to a notionally common scale.

```python
from IPython.display import Image
Image("DataNormalization.png")
```

In simple terms, Normalization helps to bring in all features in the range of [0-1]

Here income is about 1000 times than range.
So these two features are in very different ranges.
When we do further analysis, like linear regression, for example, the attribute “income” will
intrinsically influence the result more, due to its larger value, but this doesn’t necessarily
mean it is more ‘important’ as a predictor.
So, the nature of the data biases the linear regression model to weigh income more heavily
than age.
To avoid this, we can normalize these two variables into values that range from 0 to 1.


Methods of normalizing data:
1.	Simple Feature scaling: Just divide each value by maximum value of that feature. 
This makes new values to range b/w 0-1
2.	Min-Max: Take each value of that feature, subtract it from the minm value of that feature and divide it by range of that feature.
This makes new values to range b/w 0-1

# Standardaziation

Standardization helps to transform data so that it has zero mean and unit variance.

```python
Image("Standardization-Zscore.png")
```

Z Score or Standard Score: Subtract each value by mean of the feature and then divide it by standard deviation.

__The resulting values typically range between -3 and +3, but can be higher or lower__


Use function __sklearn.preprocessing.scale()__

Parameters:
 X: Data to be scaled

with_mean: Boolean. Whether to center the data (make zero mean)

with_std: Boolean (whether to make unit standard deviation


__Standardization__ transforms your data such that the resulting distribution has a mean of 0 and a standard deviation of 1

Normalization/standardization are designed to achieve a similar goal, which is to create features that have similar ranges to each other. We want that so we can be sure we are capturing the true information in a feature, and that we dont over weigh a particular feature just because its values are much larger than other features.

If you're going to be normalizing at least one variable/feature, I would do the same thing to all of the others as well

# CODE FOR NORMALIZATION

```python
from sklearn import preprocessing
import numpy as np
data = np.array([[ 1., -1.,  2.],
             [ 2.,  0.,  0.],
             [ 0.,  1., -1.]])

data_normalized = preprocessing.normalize(data)

                                         

```

# Binning / Discretization

 - Converting a number of occurances of numeric values to few catagories.
 
 For E.g: 0-2 Childerns Green Card and so on
 * The process of grouping values together into bins
* Binning converts numeric to categorical variables.
It groups a set of numerical values into a set of bins
* Sometimes binning is helpful in increasing the accuracy of the predictive models

_For example, you can bin “age” into [0 to 5], [6 to 10], [11 to 15] and so on._


In addition, sometimes we use data binning to group a set of numerical values into a smaller number of bins to have a better understanding of the data distribution.

As example, “price” here is an attribute range from 5,000 to 45,500.Using binning, we categorize the price into three bins: low price, medium price, and high prices.


* We can use __cut and qcut functions __ available in pandas to do binning
* __EQUAL WIDTH__ : use cut()
* __EQUAL FREQUENCY__ : Use qcut()

* Orange has discretization functions in Orange.feature.discretization
