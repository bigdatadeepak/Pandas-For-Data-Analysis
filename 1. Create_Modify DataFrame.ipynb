# What kind of problem does pandas solve?
Python has long been great for data munging and preparation, but less so for data analysis and modeling. 

Pandas helps to fill this gap, enabling us to carry out our entire data analysis workflow in Python without having to switch to a more domain specific language like R.

Combined with the excellent IPython toolkit and other libraries, the environment for doing data analysis in Python excels in performance, productivity, and the ability to collaborate.

pandas does not implement significant modeling functionality outside of linear and panel regression; for this, look to statsmodels and scikit-learn. More work is still needed to make Python a first class statistical modeling environment, but we are well on our way toward that goal.

We convert datasets to python object using pandas library and we call them as data frame

DataFrame represents data in tabular structure having index over both records and attributes/features

# Insert, Update and Delete records from a dataframe

# Inserting record at the end

```python
"""
>>> df
   ColA  ColB  ColC
0     1     2     3
1     4     5     6
"""
#: Insert row(s) at the end
idx = len(df)
df.loc[idx] = [7,8,9]
df
"""
OUTPUT
--------
  ColA ColB ColC
0    1    2    3
1    4    5    6
2    7    8    9
"""
```

# Update Data Element with specific row and column

```python
"""
Update row(s) based on criterion
"""
df.loc[  df.ColB%2 == 0,  'ColC'  ] = 10
df
"""
OUTPUT
-------
   ColA  ColB  ColC
0     1     2    10
1     4     5     6
2     7     8    10

"""
```

# Delete specifc row based on business condition

```python
df.drop(labels=['row_label'], axis=0, inplace=True)
df
```

```python
"""
Here we are keeping all records where ColC == 10
"""
df = df[ df.ColC == 10  ]
df
"""
OUTPUT
-------
   ColA  ColB  ColC
0     1     2    10
2     7     8    10

"""
```

# Set Index of records of the DataFrame

```python
df = pd.DataFrame({'month': [1, 4, 7, 10],
                   'year': [2012, 2014, 2013, 2014],
                   'sale':[55, 40, 84, 31]})
```

```python
df
```

Let's set the "month" column to be index of this dataframe

```python
df.set_index('month',inplace=True)
```

```python
df
```

We Can set multiple columns for indexing as well...We can that **MULTI INDEXING**

```python
df = pd.DataFrame({'month': [1, 4, 7, 10],
                   'year': [2012, 2014, 2013, 2014],
                   'sale':[55, 40, 84, 31]})
df
```

```python
df.set_index(['year', 'month'],inplace=True)
#df.set_index([[1, 2, 3, 4], 'year'],inplace=True)
df#here 'year' and 'month' are multiple indexes
```
